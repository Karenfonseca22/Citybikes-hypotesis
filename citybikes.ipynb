```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Reemplaza con tu ID de proyecto y el nombre de la tabla
project_id = 'reto-tecnico-431721'  # Asegúrate de que este sea tu ID de proyecto real
dataset_id = 'citybikes'  # Nombre del conjunto de datos
table_id1 = 'version_1'  # Nombre de la tabla 1
table_id2 = 'citybikes_final'  # Nombre de la tabla 2

# Consulta SQL para seleccionar los datos de la tabla
query1 = f'SELECT * FROM `{project_id}.{dataset_id}.{table_id1}`'  # Corrección en la construcción de la consulta
query2 = f'SELECT * FROM `{project_id}.{dataset_id}.{table_id2}`'  # Corrección en la construcción de la consulta

df_versionu = pd.read_gbq(query1, project_id = project_id)
df_citybikes = pd.read_gbq(query2, project_id = project_id)

# Mostrar las primeras filas del DataFrame
display(df_versionu)
display(df_citybikes)
```

Boxplot de stoptime_date

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(10, 6))

# Convert dates to matplotlib datenums for plotting
dates = mdates.date2num(df_versionu['stoptime_date'])
plt.boxplot(dates, patch_artist=True, notch=True)

# Format x-axis to display dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # Rotate date labels for better readability

plt.title('Boxplot de stoptime_date')
plt.xlabel('stoptime_date')
plt.ylabel('cantidad')
plt.grid(True)
plt.show()
```

Viajes promedio en un dia

```python
import pandas as pd

# Dataframe llamado df_citybikes
df_citybikes['stoptime_date'] = pd.to_datetime(df_citybikes['stoptime_date'])  # Asegurarse de que las fechas estén en formato datetime

# Agrupar por día y contar los viajes
viajes_por_dia = df_citybikes.groupby(df_citybikes['stoptime_date'].dt.date).size()

# Calcular el promedio de viajes por día
promedio_viajes_diarios = viajes_por_dia.mean()

print(f"El número promedio de viajes por día es: {promedio_viajes_diarios}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear la gráfica de línea con seaborn
plt.figure(figsize=(10,6))
sns.lineplot(x=viajes_por_dia.index, y=viajes_por_dia.values, marker='o', color='blue')
plt.title('Número de viajes por día')
plt.xlabel('Fecha')
plt.ylabel('Número de viajes')
plt.xticks(rotation=45)  # Rotar etiquetas de fechas para mejor visibilidad
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tu DataFrame se llama 'df_citybikes' y la columna de fecha es 'stoptime_date'

# Convertir la columna 'stoptime_date' a formato datetime si no lo está
df_citybikes['stoptime_date'] = pd.to_datetime(df_citybikes['stoptime_date'])

# Agrupar por día y contar el número de viajes
viajes_diarios = df_citybikes.groupby(df_citybikes['stoptime_date'].dt.date).size()

# Calcular la tasa de crecimiento diaria (opcional)
crecimiento_diario = viajes_diarios.pct_change().fillna(0) * 100  # Porcentaje de cambio

# Crear la gráfica de número de viajes diarios
plt.figure(figsize=(12, 6))

# Gráfico del número de viajes diarios
plt.subplot(2, 1, 1)
plt.plot(viajes_diarios.index, viajes_diarios.values, marker='o', color='blue')
plt.title('Número de Viajes Diarios')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.grid(True)
plt.xticks(rotation=45)  # Rotar etiquetas de fechas para mejor visibilidad

# Gráfico del crecimiento diario (opcional)
plt.subplot(2, 1, 2)
plt.plot(viajes_diarios.index, crecimiento_diario, marker='o', color='green')
plt.title('Tasa de Crecimiento Diario (%)')
plt.xlabel('Fecha')
plt.ylabel('Tasa de Crecimiento (%)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotar etiquetas de fechas para mejor visibilidad

plt.tight_layout()
plt.show()
```
